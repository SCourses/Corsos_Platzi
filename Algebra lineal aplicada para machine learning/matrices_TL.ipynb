```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
matriz=np.array([[-1,3],[2,-2]])
vector=np.array([[2],[1]])
```

```python
def graficarVectores(vecs, cols, alpha = 1):
    
   
    plt.axvline(x = 0, color = "grey", zorder = 0)
    plt.axhline(y = 0, color = "grey", zorder = 0)
    
    for i in range(len(vecs)):
        x = np.concatenate([[0,0], vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                  angles='xy', scale_units='xy', scale=1,
                   color=cols[i],
                  alpha = alpha)
```

```python
graficarVectores([vector.flatten()],cols="blue")
plt.xlim(-0.5,3)
plt.ylim(-0.2,2)
```

```python
vector_transformado=matriz.dot(vector)
graficarVectores([vector.flatten(),vector_transformado.flatten()],cols=["blue","orange"])
plt.xlim(-0.5,2.5)
plt.ylim(-0.2,2.5)
print(vector_transformado)
```

## Tres transformaciónes


```python
def graficarMatriz(matriz,vectorCol=["red","blue"]):
    #circulo unitario 
    x=np.linspace(-1,2,100000)
    y=np.sqrt(1-(x**2))
    #curculo unitario transformado
    x1=matriz[0,0]*x+matriz[0,1]*y
    y1=matriz[1,0]*x+matriz[1,1]*y
    x1_neg=matriz[0,0]*x-matriz[0,1]*y
    y1_neg=matriz[1,0]*x-matriz[1,1]*y

    #vectores
    u1=[matriz[0,0],matriz[1,0]]
    v1=[matriz[0,1],matriz[1,1]]

    graficarVectores([u1,v1],cols=[vectorCol[0],vectorCol[1]])
    plt.plot(x1,y1,"green",alpha=0.7)
    plt.plot(x1_neg,y1_neg,"green",alpha=0.7)
```

```python
A=np.array([[3,7],[5,2]])
print(A)
```

```python
print("Circulo unitario:")
graficarMatriz(np.array([[1,0],[0,1]]))
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
print("Circulo unitario transformado:")
graficarMatriz(A)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

## D V U y su efecto en la transformación

```python
print("Escala(D):")
graficarMatriz(np.array(np.diag(D).dot(V)))
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()
```

## Pseudoinversa de Moore Penrose

```python
np.set_printoptions(suppress=True) ## todos los números cercanos a cero sean 0
```

```python
A=np.array([[2,3],[5,7],[11,13]])
print(A)
```

```python
U, D , V=np.linalg.svd(A)
```

```python
D_pse=np.zeros((A.shape[0],A.shape[1])).T
print(D_pse)
```

```python
print("Valores a reemplazar en D_pse")
print(D_pse[:D.shape[0],:D.shape[0]])
```

```python
print("Valores que pondremos en D_pse")
print(np.linalg.inv(np.diag(D)))
```

```python
D_pse[:D.shape[0],:D.shape[0]]=np.linalg.inv(np.diag(D))
```

```python
D_pse
```

```python
A_pse=V.T.dot(D_pse).dot(U.T)
A_pse
```

```python
A_pse_calc = np.linalg.pinv(A)
A_pse_calc
```

## Preparando el conjunto de imágenes para aplicar PCA


```python
import imageio
im=imageio.imread("imagen_ejemplo_frida_bredesen.jpg")
im=im.astype(np.uint8)
```

```python
#Normalizamos nustra imágen 
im_original = im
im=im/255
```

```python
fix,ax=plt.subplots(1,2,figsize=(12,12),subplot_kw={"xticks":[],"yticks":[]})
ax[0].imshow(im_original,cmap="gray")
ax[1].imshow(im,cmap="gray")
```

```python
from glob import iglob
import pandas as pd

caras = pd.DataFrame([])

for path in iglob("imagenes/*/*.pgm"):
    im=imageio.imread(path)
    cara=pd.Series(im.flatten(),name=path)
    caras=caras.append(cara)

fix,axes=plt.subplots(5,10,figsize=(15,8),subplot_kw={"xticks":[],"yticks":[]},gridspec_kw=dict(hspace=0.01,wspace=0.01))
for i, ax in enumerate(axes.flat):
    ax.imshow(caras.iloc[i].values.reshape(112,92),cmap="gray")
```

```python
from sklearn.decomposition import PCA
```

```python
caras_pca=PCA(n_components=0.8)
caras_pca.fit(caras)
```

```python
filas=3
columnas=caras_pca.n_components_//filas

fig,axes=plt.subplots(filas,columnas,figsize=(12,6),subplot_kw={"xticks":[],"yticks":[]},gridspec_kw=dict(hspace=0.01,wspace=0.01))
for i, ax in enumerate(axes.flat):
    ax.imshow(caras_pca.components_[i].reshape(112,92),cmap="gray")
```

```python
print(caras_pca.n_components_)

componentes=caras_pca.transform(caras)
proyeccion=caras_pca.inverse_transform(componentes)
```

```python
fig,axes=plt.subplots(5,10,figsize=(15,8),subplot_kw={"xticks":[],"yticks":[]},gridspec_kw=dict(hspace=0.01,wspace=0.01))
for i, ax in enumerate(axes.flat):
    ax.imshow(proyeccion[i].reshape(112,92),cmap="gray")
```
