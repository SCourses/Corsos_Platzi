# ***Copy***

```python
arr=np.arange(0,11)
arr
```

```python
trozo_de_arr=arr[0:6]
trozo_de_arr
```

```python
trozo_de_arr[:]=0
print(trozo_de_arr)
print(arr) #se afecta el objeto padre
```

```python
arr_copy=arr.copy() #se respeta el valor de lalista original
arr_copy
```

```python
arr_copy[:]=100
print(arr_copy)
print(arr) #prueba
```

# ***Condiciones***

```python
arr=np.linspace(1,10,10,dtype='int8')
arr
```

```python
indices_cond=arr>5
indices_cond
```

```python
arr[indices_cond] #filtra los que estan en true
arr[arr>5] #filtra los que estan en true
```

```python
arr[(arr>5) & (arr<9)]
```
