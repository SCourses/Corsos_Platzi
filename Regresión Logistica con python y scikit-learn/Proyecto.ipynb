```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
df=pd.read_csv("data.csv")
df_data=df.copy()
df
```

```python
df.info()
```

```python
df.diagnosis.unique()
```

```python
df.diagnosis.replace(["M","B"],[1,0],inplace=True)
```

```python
df.drop(columns=["id","Unnamed: 32"],inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),annot=True)
```

```python
df.drop(columns=["perimeter_mean"],inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_data_processing_scaled =  scaler.fit_transform(df)
df_data_processing_scaled = pd.DataFrame(df_data_processing_scaled)
df_data_processing_scaled.columns=df.columns
```

```python
X=df_data_processing_scaled.drop("diagnosis",axis=1)
y=df_data_processing_scaled["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
from sklearn import metrics
prediction_test = model.predict(X_test)
print(metrics.accuracy_score(y_test,prediction_test ))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
fig = plt.figure(figsize=(11,11))
cm = confusion_matrix(y_test, prediction_test, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=model.classes_)
disp.plot(cmap='gray')
plt.show()
```
