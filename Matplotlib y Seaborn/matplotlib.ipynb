# ***SubPlot***

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b--')
plt.subplot(2,1,2)
plt.hist(y)
plt.show()
```

# ***Mtodo Orientado a Objeto***

```python
fig=plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
fig.show()
```

```python
fig=plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])
axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')
fig.show()
```

# ***Subplots***

```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

```python
fig,axes=plt.subplots()
axes.plot(x,y,'b')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig,(axes1,axes2)=plt.subplots(nrows=1,ncols=2)
axes1.plot(x,y,'b')
axes2.plot(y,x,'r')
```

```python
fig,axes=plt.subplots(2,4)
axes[0,0].plot(x,y,'b')
axes[0,1].plot(y,x,'r')
axes[0,2].plot(y,x,'r')
axes[0,3].plot(y,x,'r')
axes[1,0].plot(y,np.cos(x),'r')
```

# ***Leyendas,Etiquetas,Titulos,tamaos***

```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,y,label="Sin(X)")
axes[0].set_title('Relacin X-Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y,x)
axes[1].set_title('Relacin Y-X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x,y,label="sin(x)")
plt.title('hola q hace')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

# ***Colores y estilos***

```python
x=np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
fig,ax=plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'bo')
ax.plot(x,x+3,'yD')
ax.plot(x,x+4,'g')
```

```python
plt.style.use('dark_background')
fig,ax=plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='red')
ax.plot(x,x+2,'bo')
ax.plot(x,x+3,'yD')
ax.plot(x,x+4,'g')
```

# ***Barplot***

```python
countrys = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", 'GERMANY']
population = [5000, 800, 900, 1000, 300]
```

```python
plt.style.use('Solarize_Light2')
plt.barh(countrys,population)
plt.show()
```

```python
plt.style.use('Solarize_Light2')
plt.bar(countrys,population,width=0.5,color=['red','blue'])
plt.show()
```
