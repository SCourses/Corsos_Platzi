# ***CategÃ³ricas***

```python
tips.head()
```

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show() #concentraciÃ³n de los datos
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,color='black')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='box')
plt.show() #cualquier tipo de grafico con kind
```

# ***Relations***

```python
tips.head(2)
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
markers={"Lunch":"D","Dinner":"s"}
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers,kind='line',col='time')
plt.show()
```

# ***Joinplot/Pairplot***

```python
tips.head()
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='hist',marginal_ticks=True,marginal_kws=dict(bins=25,fill=False,multiple='dodge'))
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',corner=True)
plt.show() #relaciÃ³n entre las variables numÃ©ricas
```
