## 2) Muestreo  Sistemático


```python
def systematic_sampling(df,steps):
    indexes=np.arange(0,len(df),step=steps)
    systematic_sample = df.iloc[indexes]
    return systematic_sample

systematic_sampling(df,2)
```

## Muestreo estratificado


```python
df["Estratificado"]=df["delegacion"]+", "+df["tipo"]

(df["Estratificado"].value_counts()/len(df)).sort_values(ascending=False)
```

Nuestros datos dicen que la proporción es la siguiente 

1)Hoteles  en cautemoc 0.5

2)Cuautémoc,Museo                0.2

3)Venustiano Carranza,Hotel      0.1

4)Cuauhtémoc,Mercado             0.1

5)Venustiano Carranza,Mercado    0.1

```python
def data_estratificada(df,nombres_columnas_estrat,valores_estrat,prop_estrat,random_state=None):
    df_estra = pd.DataFrame(columns=df.columns)

    pos=-1
    for i in range(len(valores_estrat)):
        pos +=1
        if pos == len(valores_estrat)- 1:
            ratio:len=len(df)-len(df_estra)
        else:
            ratio_len=int(len(df)*prop_estrat[i])

        df_filtrado=df[df[nombres_columnas_estrat]==valores_estrat[i]]
        df_temp=df_filtrado.sample(replace=True,n=ratio_len,random_state=random_state)

        df_estra=pd.concat([df_estra,df_temp])
    return df_estra
    
```

```python
valores_estrat=["Cuautémoc, Hotel","Cuautémoc, Museo","Venustiano Carranza, Hotel","Cuauhtémoc, Mercado","Venustiano Carranza, Mercado"]
prop_estrat =[0.5,0.2,0.1,0.1,0.1]
df_estratificada= data_estratificada(df,"Estratificado",valores_estrat,prop_estrat,random_state=42)
```

```python
(df_estratificada["Estratificado"].value_counts()/len(df)).sort_values(ascending=False)
```
